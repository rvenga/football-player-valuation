```python
# for table analysis
import pandas as pd

# write to csv
import csv
```

```python
def League_table_scraper(url,position_of_table_in_the_page):
    """
    Function takes in the the url of the wikipedia page of 
    a leagues season and the position of the table in the 
    webpage as arguments and returns the league table 
    for that season
    
    Arg:
    string: league season url
    int : position of the league table 
    (best approach is to try number 3-10 to find the right table)
    
    Out:
    dataframe of leagur table
    
    """
    
    # Getting the data
    
    df = pd.read_html(url)[position_of_table_in_the_page]
    df.to_csv('League_table.csv',index=False)
    
    
    # Cleaning the data
    
    # Fix column names
    df.rename(columns={ df.columns[1]: "club" }, inplace = True)
    
    # Clean team names
    teams = list(df.club)
    # Iterate over the list of teams
    for i, team in enumerate(teams):
        # Split the string at the first occurrence of '(' or '['
        split_team = team.split('(',1)[0].split('[', 1)[0]
        # Get the first part of the split string (the team name) and strip any leading or trailing white space
        cleaned_team = split_team.strip()
        # Update the element in the list with the cleaned team name
        teams[i] = cleaned_team

    df['club'] = teams
    
    return df
```

```python
League_table_scraper('https://en.wikipedia.org/wiki/2019%E2%80%9320_Belgian_First_Division_A',4)
```
