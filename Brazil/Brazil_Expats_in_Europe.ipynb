#### Mapping function for club balance

```python
def balance_map(row):
    """
    function takes in each row as argument and
    return 1 if the club of the row has a negative
    transfer balance, 0 if it has a non-negative 
    balance and returns -1 if the clubs 
    doesn't exist in our data base
    
    """
    club = row['club_code']
    club_balance = df_balance[df_balance['club_code'] == club]
        
    if  (club_balance['balance_usd'] < 0).any():
        return 1
    if  (club_balance['balance_usd'] >= 0).all():
        return 0
    else:
        return -1
```

#####  Exaction Function

```python
def Age(age):
    """
    Function takes age as an argument and returns 
    the age range
    
    """
    if age <= 18:
        return 'U18'
    if age in range(18, 22):
        return '18-21'
    if age in range(22, 24):
        return '21-23'
    if age in range(24, 28):
        return '23-27'
    if age in range(28, 33):
        return '27-32'
    return '32+'
```

```python
def expat_players(age, position, starter, negative_balance, lower_wagebound, upper_wagebound):
    """
    Function takes age, position, starter or not,negative club balance or not  and the lower and upper bounds 
    of wage and returns a dataframe containing the players from the top 6 leagues 
    matching the criteria.
    """
    df_expat = pd.DataFrame()

    df_expat = df_brazil[df_brazil['player_group_code'] == str(position)]

    df_expat = df_expat[df_expat['Age'] == Age(age)]

    df_expat = df_expat[df_expat['high_earner'] == starter]
    
    df_expat = df_expat[df_expat['negative_balance'] == negative_balance]
    
    df_expat = df_expat[df_expat['salary_gross_eur'].between(lower_wagebound, upper_wagebound)]
    
    df_expat = df_expat.drop(columns=['club_code','player_age'])
    
    return df_expat
```
